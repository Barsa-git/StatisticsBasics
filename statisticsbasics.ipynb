**2**.What are the two main types of statistics?
 - The two main types of statistics are descriptive statistics, which summarize and describe data, and inferential statistics, which use data to make inferences or predictions about a population.

**3**.What are descriptive statistics?
 - Descriptive statistics summarize and describe the key features of a dataset, such as its central tendency, variability, and distribution, without making inferences about a larger population.

**4**.What is inferential statistics?
 - Inferential statistics is a branch of statistics that allows you to draw conclusions and make predictions about a population based on a sample of data, using techniques like hypothesis testing and confidence intervals.

**5**.What is sampling in statistics?
 - In statistics, sampling is the process of selecting a subset (a sample) from a larger population to make inferences or draw conclusions about the entire population, saving time and resources while still obtaining meaningful insights.

**6**.What are the different types of sampling methods?
 - Sampling methods are broadly categorized into probability (random) and non-probability (non-random) sampling, each with various techniques like simple random, stratified, systematic, cluster, and convenience sampling, among others.
 - Probability (Random) Sampling : -
  - Simple Random Sampling : Every member of the population has an equal chance of being selected.
  - Stratified Sampling : The population is divided into subgroups (strata), and then a random sample is taken from each stratum.
  - Systematic Sampling : Elements are selected at regular intervals from a list or population.
  - Cluster Sampling : The population is divided into clusters, and then a random sample of clusters is selected.
  - Multistage Sampling : Combines different probability sampling methods at different stages of research.
 - Non-Probability (Non-Random) Sampling : -
  - Convenience Sampling : Participants are selected based on ease of access.
  - Snowball Sampling : Participants are recruited through referrals from other participants.
  - Quota Sampling : Participants are selected to match specific characteristics or quotas within the population.
  - Purposive Sampling : Participants are selected based on specific criteria or characteristics relevant to the research.  

**7**.What is the difference between random and non-random sampling?
 - Random sampling (also known as probability sampling) ensures every member of the population has an equal chance of being selected, while non-random sampling (non-probability sampling) relies on factors other than chance, potentially leading to biased results.

**14**.What is variance, and how is it calculated?
 - Variance is a statistical measurement of how spread out data points are in a data set. It's calculated by finding the average of the squared differences between each data point and the mean.
 - How to calculate variance : -
  - Calculate the mean of the data set
  - Subtract the mean from each data point
  - Square each of the differences
  - Add up all of the squared differences
  - Divide the sum of the squared differences by the number of data points minus one
 - Types of variance : -
  - Population variance : Usually denoted by σ2
  - Sample variance : Usually denoted by s2

**17**.What is the difference between variance and standard deviation?
 - Variance measures the average squared difference of each data point from the mean, while standard deviation is the square root of the variance, representing the average distance of data points from the mean in the original data units.
 - Variance :
   - Measures how much data points deviate from the mean.
   - Calculated as the average of the squared differences between each data point and the mean.
   - Expressed in squared units (e.g., meters squared).
   - Can be difficult to interpret directly because it's in squared units.
   - Formula: Variance = Σ( (xᵢ - μ)² ) / N, where xᵢ is each data point, μ is the mean, and N is the number of data points.
 - Standard Deviation :
   - Represents the average spread or dispersion of data points around the mean.
   - Calculated as the square root of the variance.
   - Expressed in the same units as the original data (e.g., meters).
   - Easier to interpret than variance because it's in the same units as the data.
   - Formula: Standard Deviation = √( Variance ) = √[ Σ( (xᵢ - μ)² ) / N ].   

**18**.What is skewness in a dataset?

**19**.What does it mean if a dataset is positively or negatively skewed?
 - A positively skewed dataset has a longer tail on the right side, meaning most values cluster towards the lower end and a few high values pull the distribution's tail to the right.
 - Conversely, a negatively skewed dataset has a longer tail on the left, with most values clustering towards the higher end and a few low values pulling the distribution's tail to the left.
 - Positively Skewed (Right Skewed) : -
  - The tail of the distribution extends towards the higher values (right side).
  - Most data points are clustered towards the lower values (left side).
  - The mean is typically greater than the median.
  - Examples : Income distribution (where a few high earners pull the average income up), test scores on a difficult exam.
 - Negatively Skewed (Left Skewed) : -
  - The tail of the distribution extends towards the lower values (left side).
  - Most data points are clustered towards the higher values (right side).
  - The mean is typically less than the median.
  - Examples : Housing prices (where a few very low-priced houses pull the average price down), age of death (where a few very old people pull the average age of death up).

**20**.Define and explain kurtosis.

Distributions with medium kurtosis (medium tails) are mesokurtic. Distributions with low kurtosis (thin tails) are platykurtic.

**1**.How do you calculate the mean, median, and mode of a dataset?

```python
import numpy as np
from scipy import stats

def calculate_mean_median_mode(dataset):

  mean = np.mean(dataset)
  median = np.median(dataset)

  mode = stats.mode(dataset).mode
  if isinstance(mode, np.ndarray):
    mode = mode[0]

  return mean, median, mode

dataset = [1, 2, 3, 4, 5, 5, 6, 7, 8, 9]
mean, median, mode = calculate_mean_median_mode(dataset)
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

**4**.Implement sampling techniques like random sampling and stratified sampling.

 - Random Sampling : -

```python
import random

def simple_random_sampling(population, sample_size):

    if sample_size > len(population):
        raise ValueError("Sample size cannot exceed population size.")

    return random.sample(population, sample_size)

population = list(range(1, 101))
sample_size = 20
sample = simple_random_sampling(population, sample_size)
print(f"Simple Random Sample: {sample}")
```

 - Stratified Sampling : -

```python
import pandas as pd

def stratified_sampling(data, column_to_stratify, sample_size_per_stratum):

    strata = data.groupby(column_to_stratify)
    sampled_data = strata.apply(lambda x: x.sample(sample_size_per_stratum))
    return sampled_data

data = pd.DataFrame({
    'Age': [18, 20, 22, 25, 28, 30, 32, 35, 38, 40],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'City': ['Sukinda', 'Sukinda', 'Sukinda', 'Sukinda', 'Sukinda', 'Sukinda', 'Sukinda', 'Sukinda', 'Sukinda', 'Sukinda']})

sample_size_per_stratum = 3
stratified_sample = stratified_sampling(data, 'Gender', sample_size_per_stratum)
print(f"Stratified Sample: \n{stratified_sample}")
```

**9**.Write a Python script to calculate covariance between two datasets.

```python
import numpy as np

def calculate_covariance(x, y):

    if len(x) != len(y):
        raise ValueError("Datasets must have the same length")

    n = len(x)
    mean_x = sum(x) / n
    mean_y = sum(y) / n

    covariance = sum((x[i] - mean_x) * (y[i] - mean_y) for i in range(n)) / (n - 1)
    return covariance

dataset1 = [2, 4, 6, 8, 10]
dataset2 = [1, 3, 5, 7, 9]

covariance_manual = calculate_covariance(dataset1, dataset2)
print("Covariance (manual calculation):", covariance_manual)

covariance_numpy = np.cov(dataset1, dataset2)[0, 1]
print("Covariance (NumPy):", covariance_numpy)
```

**10**.Write a Python script to calculate the correlation coefficient between two datasets.

```python
import numpy as np

def calculate_correlation_coefficient(dataset1, dataset2):

    if not dataset1 or not dataset2:
        return None
    if len(dataset1) != len(dataset2):
        raise ValueError("Datasets must have the same length")

    dataset1 = np.array(dataset1)
    dataset2 = np.array(dataset2)

    if np.any(np.isnan(dataset1)) or np.any(np.isnan(dataset2)):
      raise ValueError("Datasets cannot contain NaN values")

    if len(dataset1) < 2:
        return None

    mean_x = np.mean(dataset1)
    mean_y = np.mean(dataset2)

    numerator = np.sum((dataset1 - mean_x) * (dataset2 - mean_y))
    denominator = np.sqrt(np.sum((dataset1 - mean_x)**2) * np.sum((dataset2 - mean_y)**2))

    if denominator == 0:
        return 0

    correlation_coefficient = numerator / denominator
    return correlation_coefficient

if __name__ == '__main__':
    data1 = [1, 2, 3, 4, 5]
    data2 = [2, 4, 1, 3, 5]

    try:
        correlation = calculate_correlation_coefficient(data1, data2)
        print(f"The correlation coefficient is: {correlation}")
    except ValueError as e:
        print(f"Error: {e}")

    data3 = [1, 2, 3]
    data4 = [4, 5]

    try:
        correlation = calculate_correlation_coefficient(data3, data4)
        print(f"The correlation coefficient is: {correlation}")
    except ValueError as e:
        print(f"Error: {e}")
```

**12**.Implement and compare simple random sampling and systematic sampling.

```python
import random

def simple_random_sampling(population, sample_size):

  if sample_size > len(population):
    raise ValueError("Sample size cannot exceed population size.")
  return random.sample(population, sample_size)


def systematic_sampling(population, sample_size):

  if sample_size > len(population):
    raise ValueError("Sample size cannot exceed population size.")
  interval = len(population) // sample_size
  start = random.randint(0, interval)
  sample = [population[i] for i in range(start, len(population), interval)]
  return sample[:sample_size]

population = list(range(1, 101))
sample_size = 10

random_sample = simple_random_sampling(population, sample_size)
systematic_sample = systematic_sampling(population, sample_size)

print("Simple Random Sample:", random_sample)
print("Systematic Sample:", systematic_sample)
```

**13**.Calculate the mean, median, and mode of grouped data.

```python
import pandas as pd

def calculate_mean_median_mode_grouped(data):

  mean = (data['midpoint'] * data['frequency']).sum() / data['frequency'].sum()

  cumulative_frequency = data['frequency'].cumsum()
  total_frequency = data['frequency'].sum()
  median_position = (total_frequency + 1) / 2
  for index, row in data.iterrows():
    if cumulative_frequency[index] >= median_position:
      median = row['midpoint']
      break

  mode_frequency = data['frequency'].max()
  mode = data[data['frequency'] == mode_frequency]['midpoint'].iloc[0]

  return mean, median, mode

data = {
    'midpoint': [10, 20, 30, 40, 50],
    'frequency': [5, 10, 15, 8, 2]
}
df = pd.DataFrame(data)
mean, median, mode = calculate_mean_median_mode_grouped(df)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
```

**18**.Implement Z-score normalization and explain its significance.

```python
import pandas as pd
import numpy as np

data = {'feature1': [10, 20, 30, 40, 50],
        'feature2': [100, 200, 300, 400, 500],
        'feature3': [1, 2, 3, 4, 5]}

df = pd.DataFrame(data)

def zscore_normalize(df):
    for column in df.columns:
        df[column] = (df[column] - df[column].mean()) / df[column].std()
    return df

normalized_df = zscore_normalize(df.copy())

print("Original DataFrame:")
print(df)

print("\nNormalized DataFrame:")
print(normalized_df)
```

**20**.Write a Python program to visualize covariance using a heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {'A': np.random.rand(100),
        'B': np.random.rand(100),
        'C': np.random.rand(100)}
df = pd.DataFrame(data)

covariance_matrix = df.cov()

plt.figure(figsize=(8, 6))
sns.heatmap(covariance_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Covariance Matrix Heatmap')
plt.show()
```

**22**.Generate a dataset and implement both variance and standard deviation computations.

```python
import statistics

data = [5, 8, 2, 10, 7, 9, 3, 6]

mean = statistics.mean(data)

def calculate_variance(data, mean):
    squared_diff = [(x - mean)**2 for x in data]
    variance = sum(squared_diff) / len(data)
    return variance

variance = calculate_variance(data, mean)

std_dev = statistics.stdev(data)

print("Dataset:", data)
print("Mean:", mean)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

**23**.Visualize skewness and kurtosis using Python libraries like matplotlib or seaborn.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import skew, kurtosis, norm

np.random.seed(42)
data_normal = np.random.normal(0, 1, 1000)
data_skewed_right = np.random.exponential(1, 1000)
data_skewed_left = -np.random.exponential(1, 1000)
data_high_kurtosis = np.concatenate([np.random.normal(-3, 0.5, 250), np.random.normal(3, 0.5, 250), np.random.normal(0, 1, 500)])

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

sns.histplot(data_normal, kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title(f'Normal Distribution (Skew: {skew(data_normal):.2f}, Kurtosis: {kurtosis(data_normal):.2f})')

sns.histplot(data_skewed_right, kde=True, ax=axes[0, 1], color='lightcoral')
axes[0, 1].set_title(f'Right-Skewed Distribution (Skew: {skew(data_skewed_right):.2f}, Kurtosis: {kurtosis(data_skewed_right):.2f})')

sns.histplot(data_skewed_left, kde=True, ax=axes[1, 0], color='lightgreen')
axes[1, 0].set_title(f'Left-Skewed Distribution (Skew: {skew(data_skewed_left):.2f}, Kurtosis: {kurtosis(data_skewed_left):.2f})')

sns.histplot(data_high_kurtosis, kde=True, ax=axes[1, 1], color='gold')
axes[1, 1].set_title(f'High Kurtosis Distribution (Skew: {skew(data_high_kurtosis):.2f}, Kurtosis: {kurtosis(data_high_kurtosis):.2f})')

plt.tight_layout()
plt.show()
```

**24**.Implement the Pearson and Spearman correlation coefficients for a dataset.

```python
import numpy as np
from scipy.stats import pearsonr, spearmanr

def calculate_pearson_spearman_correlation(data1, data2):

  pearson_corr, _ = pearsonr(data1, data2)

  spearman_corr, _ = spearmanr(data1, data2)

  return pearson_corr, spearman_corr

data1 = [1, 2, 3, 4, 5]
data2 = [2, 4, 1, 3, 5]

pearson_corr, spearman_corr = calculate_pearson_spearman_correlation(data1, data2)

print(f"Pearson correlation: {pearson_corr}")
print(f"Spearman correlation: {spearman_corr}")
```
